# Analysis of the Autism Diagnosis Dataset Based on DSM-5 Questionnaire

---

## 1. Overview of the Questionnaire and Its Purpose

- **Main Objective:**  
  The primary goal is to collect data related to the diagnosis and classification of Autism using a questionnaire based on DSM-5 criteria. This questionnaire gathers information regarding the childs condition through responses provided by parents, caregivers, or professionals. It also aims to collect data on different age groups and behaviors associated with Autism, in order to develop a machine learning model for Autism classification.

- **Target Audience:**  
  The questionnaire targets individuals from various categories such as parents, caregivers, researchers, medical practitioners, and special needs educators. This diversity helps ensure that the data comes from a broad spectrum of perspectives and expertise.

- **Significance:**  
  The data is intended to be used to build a diagnostic model that can classify Autism levels. It also aids in understanding behavioral patterns and key characteristics associated with Autism, ultimately contributing to early diagnosis and intervention.

---

## 2. Analysis of the Questionnaire Structure and Questions

### A. Types of Questions and Information Collected

1. **Demographic and Personal Information:**
   - **Question 1:** Choose a category you belong to (e.g., Parent, Caregiver, Researcher, Medical Practitioner, Special Needs Educator, Other).  
     *Purpose:* Identify the source of the information and the level of expertise.
   - **Question 2:** Are you Nigerian? (Yes/No)  
     *Purpose:* To determine the geographical and cultural background of the respondents.
   - **Question 3:** Which age group does your child belong to? (0-3, 4-6, 7-9, 10-12, 12 and above)  
     *Purpose:* Classify the data according to the childs age.
   - **Question 4:** Sex of the child (Male/Female)  
     *Purpose:* Assess if there is any gender-related influence in Autism diagnosis.
   - **Question 5:** Country of residence (Nigeria/Other)  
     *Purpose:* Compare results between respondents residing in Nigeria and those living elsewhere.
   - **Question 6:** Age of the Mother when the child was born (below 20, 20-25, 26-34, above 35)  
     *Purpose:* Investigate potential relationships between maternal age and Autism diagnosis.

2. **Diagnosis-Related Information:**
   - **Question 7:** Is your child previously diagnosed? (Yes/No)  
     *Purpose:* Identify cases with a prior diagnosis.
   - **Question 8:** If diagnosed, what was the diagnosis? (DSM-4 Diagnosis: Autism, Aspergers, PDD NOS, Other)  
     *Purpose:* Compare the previous DSM-4 diagnosis with the current criteria.
   - **Question 9:** If diagnosed, what is the DSM-5 Diagnosis? (Level 1: Mild Autism, Level 2: Moderate Autism, Level 3: Severe Autism)  
     *Purpose:* Use the DSM-5 criteria to classify the severity of Autism.
   - **Question 10:** At what age was the child diagnosed? (Options: 2, 3, 4, 5, 6, Other)  
     *Purpose:* Study the age range during which the diagnosis typically occurs.
   - **Question 11:** Where was the diagnosis performed? (Nigeria/Abroad)  
     *Purpose:* Understand the geographical distribution of the diagnosis.
   - **Question 12:** Is your child verbal, non-verbal, or has just a few words?  
     *Purpose:* Evaluate the childs verbal communication capabilities.

3. **Behavioral and Descriptive Information:**
   - **Question 13:** Provide some of your childs unique behaviors that prompted you to seek a diagnosis.  
     *Purpose:* Collect qualitative data that can be analyzed (e.g., using NLP techniques) to identify key behavioral patterns.
   - **Question 14:** Rate your childs deficits in social communication and social interaction (Mild, Moderate, Severe).  
     *Purpose:* Assess the level of social impairment.
   - **Question 15:** Rate your childs non-verbal communicative abilities (Poor, Good, Very Good).  
     *Purpose:* Evaluate the quality of non-verbal communication skills.
   - **Question 16:** Can your child develop, maintain, and understand relationships (Yes, No, Sometimes)?  
     *Purpose:* Understand social interaction and relationship-building skills.
   - **Question 17:** Does your child exhibit stereotyped or repetitive motor movements and echolalia (Yes, No, Sometimes)?  
     *Purpose:* Identify repetitive behaviors associated with Autism.
   - **Question 18:** How is your childs hyporeactivity to sensory stimuli (Mild, Moderate, Severe)?  
     *Purpose:* Evaluate the childs sensory responses to various environmental factors.

4. **Additional Information:**
   - **Question 19:** What other challenges does your child face apart from Autism? (e.g., ADHD, Epilepsy, Anxiety disorders, Specific learning difficulties, Intellectual Disability, Other developmental conditions)  
     *Purpose:* Gather data on comorbid conditions or additional developmental challenges.
   - **Final Request:** Drop your email address for potential future contact.  
     *Purpose:* Facilitate follow-up if additional information or clarification is needed.

##   Use CTGAN then Smote

```python
random_seed = 42
np.random.seed(random_seed)
random.seed(random_seed)
torch.manual_seed(random_seed)
torch.cuda.manual_seed_all(random_seed)  
torch.backends.cudnn.deterministic = True  
torch.backends.cudnn.benchmark = False  

df_original = df.copy()

X = df.drop(columns=["DSM5_Diagnosis"])
y = df["DSM5_Diagnosis"]

ctgan = CTGAN(epochs=1000, batch_size=1500)
ctgan.fit(df, discrete_columns=["DSM5_Diagnosis"])

num_samples = 10000    
synthetic_data = ctgan.sample(num_samples)



pca = PCA(n_components=2) 
original_pca = pca.fit_transform(df.drop(columns=["DSM5_Diagnosis"]))  
synthetic_pca = pca.transform(synthetic_data.drop(columns=["DSM5_Diagnosis"]))   

sns.scatterplot(x=original_pca[:, 0], y=original_pca[:, 1], label="Original", color="blue")
sns.scatterplot(x=synthetic_pca[:, 0], y=synthetic_pca[:, 1], label="Synthetic", color="orange")
plt.legend()
plt.title("PCA Comparison: Original vs Synthetic")
plt.show()

synthetic_data.to_csv("synthetic_data.csv", index=False)  
synthetic_data = pd.read_csv("synthetic_data.csv")

X_synthetic = synthetic_data.drop(columns=["DSM5_Diagnosis"])
y_synthetic = synthetic_data["DSM5_Diagnosis"]

y_counts = Counter(y_synthetic)
target_samples = max(y_counts.values())  
smote_strategy = {cls: target_samples for cls in y_counts}    
smote_enn = SMOTEENN(sampling_strategy=smote_strategy, random_state=random_seed)
X_resampled, y_resampled = smote_enn.fit_resample(X_synthetic, y_synthetic)

df_synthetic_resampled = pd.DataFrame(X_resampled, columns=X_synthetic.columns)
df_synthetic_resampled["DSM5_Diagnosis"] = y_resampled

df_synthetic_resampled = df_synthetic_resampled.drop_duplicates()



df_final = pd.concat([df_original, df_synthetic_resampled], ignore_index=True)

print("Final Dataset Shape:", df_final.shape)
print(df_final["DSM5_Diagnosis"].value_counts())

df_final.to_csv("final_dataset.csv", index=False)
```

# Modeling and Evaluating 

```python
def evaluate_model(y_test, y_pred):
    metrics = {
        'accuracy': accuracy_score(y_test, y_pred),
        'recall_macro': recall_score(y_test, y_pred, average='macro'),
        'precision_macro': precision_score(y_test, y_pred, average='macro'),
        'f1_score_macro': f1_score(y_test, y_pred, average='macro'),
        'recall_micro': recall_score(y_test, y_pred, average='micro'),
        'precision_micro': precision_score(y_test, y_pred, average='micro'),
        'f1_score_micro': f1_score(y_test, y_pred, average='micro'),
        'recall_weighted': recall_score(y_test, y_pred, average='weighted'),
        'precision_weighted': precision_score(y_test, y_pred, average='weighted'),
        'f1_score_weighted': f1_score(y_test, y_pred, average='weighted'),
    }
    return metrics
```

## LogisticRegression

```python
param_grid = {
    'C': np.logspace(-4, 0, 5),  
    'penalty': ['l2'], 
    'solver': ['liblinear']
}

cv_strategy = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    estimator=LogisticRegression(class_weight="balanced", random_state=42),
    param_grid=param_grid,
    cv=cv_strategy,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

grid_search.fit(x_train, y_train)

print("Best Parameters: ", grid_search.best_params_)

```

```python
lr = grid_search.best_estimator_

y_pred_lr_te = lr.predict(x_test)

y_pred_lr_tr = lr.predict(x_train)
```

```python
lr_metrics=evaluate_model(y_test,y_pred_lr_te)
lr_metrics
```
